```python
def stock_portfolio_tracker():
    # Hardcoded dictionary to define stock prices as per instructions
    stock_prices = {
        "AAPL": 180,
        "TSLA": 250,
        "GOOGL": 150,
        "MSFT": 400,
        "AMZN": 175
    }

    portfolio = {}
    total_value = 0

    print("--- Zaim Stock Portfolio Tracker ---")
    print(f"Available stocks: {', '.join(stock_prices.keys())}")
    
    while True:
        symbol = input("\nEnter stock symbol to add (or type 'done' to finish): ").upper()
        
        if symbol == 'DONE':
            break
        
        if symbol in stock_prices:
            try:
                quantity = int(input(f"Enter quantity for {symbol}: "))
                if quantity < 0:
                    print("Quantity cannot be negative.")
                    continue
                
                # Update portfolio and calculate value
                portfolio[symbol] = quantity
                current_price = stock_prices[symbol]
                total_value += current_price * quantity
                print(f"Added {quantity} shares of {symbol} at ${current_price} each.")
            
            except ValueError:
                print("Invalid input. Please enter a whole number for quantity.")
        else:
            print(f"Stock '{symbol}' not found in our database. Please try AAPL, TSLA, etc.")

    # Display results
    print("\n" + "="*30)
    print("YOUR PORTFOLIO SUMMARY")
    print("="*30)
    
    report_lines = ["Stock Portfolio Report\n", "----------------------\n"]
    
    for stock, qty in portfolio.items():
        subtotal = qty * stock_prices[stock]
        line = f"{stock}: {qty} shares | Value: ${subtotal}"
        print(line)
        report_lines.append(line + "\n")
    
    final_total = f"\nTotal Investment Value: ${total_value}"
    print(final_total)
    report_lines.append(final_total)

    # Save to file (optional requirement)
    try:
        with open("portfolio_report.txt", "w") as f:
            f.writelines(report_lines)
        print("\n[System] Report successfully saved to 'portfolio_report.txt'")
    except Exception as e:
        print(f"\n[Error] Could not save file: {e}")

if __name__ == "__main__":
    stock_portfolio_tracker()
```
